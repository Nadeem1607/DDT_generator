Import necessary packages

```python
import pandas as pd
```

DDT Generation function

```python
def ddt_generator(sbox):
    size = len(sbox)
    table = [[0] * size for _ in range(size)]
    for delX in range(size):
        for x1 in range(size):
            x2 = x1 ^ delX
            y1 = sbox[x1]
            y2 = sbox[x2]
            delY = (y1 ^ y2)
            table[delX][delY] += 1
    return table
```
